```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# ข้อมูลจากตาราง
time = np.array([0, 3, 6, 9, 12])  # เดือน
weight = np.array([5, 30, 100, 180, 250])  # กรัม

# สมการ von Bertalanffy
def von_bertalanffy(t, k):
    w_infinity = 300  # น้ำหนักสูงสุด (สมมุติว่า 300 กรัม)
    return w_infinity * (1 - np.exp(-k * t))**3

# ใช้ curve_fit เพื่อหาค่า k ที่เหมาะสม
popt, pcov = curve_fit(von_bertalanffy, time, weight)
k = popt[0]

# แสดงผลค่า k
print(f"ค่า k ที่คำนวณได้: {k:.4f}")

# วาดกราฟการฟิตข้อมูล
time_fit = np.linspace(0, 12, 100)
weight_fit = von_bertalanffy(time_fit, k)

plt.plot(time, weight, 'o', label='real data')
plt.plot(time_fit, weight_fit, '-', label=f'fitting (k={k:.4f})')
plt.xlabel('time (month)')
plt.ylabel('weight (g)')
plt.legend()
plt.show()
```
